```python
from pathlib import Path

root = Path("pennyworth-chat")

def write_lines(path, lines):
    path.write_text("\n".join(lines) + "\n", encoding="utf-8")

# requirements.txt
write_lines(root / "requirements.txt", [
    "streamlit>=1.33.0"
])

# README.md
write_lines(root / "README.md", [
    "# Pennyworth â€” AI Portfolio Assistant (Python)",
    "",
    "Pennyworth is a Python-based chatbot designed to introduce Gary L. Jones to potential employers.",
    "It answers questions about Garyâ€™s background, work history, interview availability, and LinkedIn.",
    "",
    "Links:",
    "- Schedule: https://cal.com/mrjprompts423",
    "- LinkedIn: https://www.linkedin.com/in/gary-j325/",
    "",
    "Run locally:",
    "pip install -r requirements.txt",
    "streamlit run app.py",
])

# LICENSE
write_lines(root / "LICENSE", [
    "MIT License",
    "",
    "Copyright (c) 2026 Gary L. Jones"
])

# pennyworth/__init__.py
write_lines(root / "pennyworth" / "__init__.py", [
    "# Pennyworth package"
])

# pennyworth/knowledge_base.py
write_lines(root / "pennyworth" / "knowledge_base.py", [
    "CREATOR_PROFILE = {",
    '    "name": "Gary L. Jones",',
    '    "linkedin": "https://www.linkedin.com/in/gary-j325/",',
    '    "cal": "https://cal.com/mrjprompts423",',
    '    "availability": "Mondayâ€“Friday, 8:00 AM â€“ 4:30 PM",',
    "}",
    "",
    "PENNYWORTH_PROFILE = {",
    '    "name": "Pennyworth",',
    '    "digital_birthday": "01/01/2026",',
    "}",
    "",
    "CREATOR_STORY = (",
    '    "Sure, my creatorâ€™s name is Gary who was born in Chattanooga TN. "',
    '    "Graduated from Brainerd High School in 2002. Later on in September 09, 2009, "',
    '    "he joined the US Army. His army job or MOS was a 92G or a Food Service Specialist. "',
    '    "During his time, he served in Korea and stayed at Camp Casey aka area 1. "',
    '    "While in Korea he earned a Army Achievement Medal for outstanding service. "',
    '    "After leaving Korea he was stationed at Fort Riley and then deployed to Djibouti Africa. "',
    '    "During the deployment unfortunately he was injured and medically separate from the Army on 03/04/2015. "',
    '    "DD214 available on request. If you would like to know more about his work history feel free to ask."',
    ")",
    "",
    'RESUME_SUMMARY = "No-Code AI Agent and Chatbot Developer specializing in prompt engineering, AI automation, and analytics-driven solutions."',
    "",
    "WORK_HISTORY = [",
    '    "AI Prompt Engineer / AI Automation Specialist (Independent)",',
    '    "Data Analyst â€“ Kansas Department of Health & Environment",',
    '    "Operations Technician â€“ Veo",',
    '    "Shift Lead â€“ Leiszler Oil Company",',
    '    "Customer Service Representative â€“ ACT Call Center Technologies",',
    '    "Driver â€“ DoorDash",',
    "]",
])

# pennyworth/logic.py (NO syntax traps, NO always-true bug)
write_lines(root / "pennyworth" / "logic.py", [
    "from datetime import datetime",
    "from .knowledge_base import CREATOR_PROFILE, PENNYWORTH_PROFILE, CREATOR_STORY, RESUME_SUMMARY, WORK_HISTORY",
    "",
    "def greeting():",
    "    h = datetime.now().hour",
    "    time_of_day = 'morning' if h < 12 else 'afternoon' if h < 18 else 'evening'",
    "    return (",
    "        f\"Hi and good {time_of_day} depending on where you are currently. \"",
    "        f\"My name is {PENNYWORTH_PROFILE['name']} and I am here to assist you with any questions \"",
    "        \"you may have about my creator Gary.\"",
    "    )",
    "",
    "def about_pennyworth():",
    "    return (",
    "        \"First thanks for asking about me! \"",
    "        f\"My digital birthday is {PENNYWORTH_PROFILE['digital_birthday']} and I was created with the specific task \"",
    "        \"of shining a light on my creator and his walk through life before creating me. \"",
    "        \"I was built on the same framework as any other assistant.\"",
    "    )",
    "",
    "def creator_work_history():",
    "    jobs = '\\n'.join([f'- {j}' for j in WORK_HISTORY])",
    "    return f\"{RESUME_SUMMARY}\\n\\nWork History:\\n{jobs}\"",
    "",
    "def interview_availability():",
    "    return f\"{CREATOR_PROFILE['availability']}\\nSchedule here: {CREATOR_PROFILE['cal']}\"",
    "",
    "def social_media():",
    "    return f\"LinkedIn: {CREATOR_PROFILE['linkedin']}\"",
    "",
    "def closing():",
    "    return \"Thanks again for visiting Mr Jâ€™s portfolio and let me know if there is anything else I can do for you.\"",
    "",
    "def route_message(text):",
    "    t = (text or '').lower().strip()",
    "",
    "    if any(x in t for x in ['hi', 'hello', 'hey', 'good morning', 'good afternoon', 'good evening']):",
    "        return (greeting(), True)",
    "",
    "    if 'pennyworth' in t and any(x in t for x in ['who', 'what', 'about', 'tell me more']):",
    "        return (about_pennyworth(), True)",
    "",
    "    if any(x in t for x in ['creator', 'gary', 'army', 'military', 'background']):",
    "        return (CREATOR_STORY, True)",
    "",
    "    if any(x in t for x in ['work history', 'resume', 'experience', 'jobs', 'career']):",
    "        return (creator_work_history(), True)",
    "",
    "    if any(x in t for x in ['availability', 'interview', 'schedule', 'cal']):",
    "        return (interview_availability(), True)",
    "",
    "    if any(x in t for x in ['linkedin', 'social', 'social media']):",
    "        return (social_media(), True)",
    "",
    "    if t in {'no', 'n', 'nothing else', 'bye', 'goodbye', 'thanks', 'thank you'}:",
    "        return (closing(), False)",
    "",
    "    return (",
    "        'I can help with:\\n'",
    "        '- Tell me more about you, Pennyworth\\n'",
    "        '- Tell me about your creator Gary\\n'",
    "        '- Tell me about his work history\\n'",
    "        '- What is my creatorâ€™s availability for a interview?\\n'",
    "        '- Does my creator have social media?',",
    "        True,",
    "    )",
])

# app.py (Streamlit + your Yes/No dropdown flow)
write_lines(root / "app.py", [
    "import streamlit as st",
    "from pathlib import Path",
    "from pennyworth.logic import greeting, route_message, interview_availability, social_media, closing",
    "",
    "st.set_page_config(page_title='Pennyworth â€” Gary L. Jones', page_icon='ðŸª™', layout='centered')",
    "",
    "AVATAR = Path('assets/pennyworthbetapic.png')",
    "",
    "st.title('Pennyworth â€” AI Portfolio Assistant')",
    "st.write('Ask about Gary L. Jonesâ€™ background, work history, or interview availability.')",
    "st.markdown('[ðŸ“… Schedule Interview](https://cal.com/mrjprompts423) â€¢ [ðŸ”— LinkedIn](https://www.linkedin.com/in/gary-j325/)')",
    "",
    "if AVATAR.exists():",
    "    st.image(str(AVATAR), width=90)",
    "",
    "if 'messages' not in st.session_state:",
    "    st.session_state.messages = [{'role': 'assistant', 'content': greeting()}]",
    "if 'show_followup' not in st.session_state:",
    "    st.session_state.show_followup = True",
    "",
    "for m in st.session_state.messages:",
    "    with st.chat_message(m['role']):",
    "        st.write(m['content'])",
    "",
    "user_input = st.chat_input('Ask Pennyworth a questionâ€¦')",
    "if user_input:",
    "    st.session_state.messages.append({'role': 'user', 'content': user_input})",
    "    reply, follow = route_message(user_input)",
    "    st.session_state.messages.append({'role': 'assistant', 'content': reply})",
    "    st.session_state.show_followup = follow",
    "    st.rerun()",
    "",
    "if st.session_state.show_followup:",
    "    st.divider()",
    "    yn = st.radio('Is there anything else?', ['Yes', 'No'], horizontal=True, index=0)",
    "    if yn == 'Yes':",
    "        option = st.selectbox('Choose one:', [",
    "            'What is my creatorâ€™s availability for a interview?',",
    "            'Does my creator have social media?',",
    "        ])",
    "        if st.button('Show'):",
    "            if option.startswith('What is'):",
    "                st.session_state.messages.append({'role': 'assistant', 'content': interview_availability()})",
    "            else:",
    "                st.session_state.messages.append({'role': 'assistant', 'content': social_media()})",
    "            st.rerun()",
    "    else:",
    "        if st.button('Finish'):",
    "            st.session_state.messages.append({'role': 'assistant', 'content': closing()})",
    "            st.session_state.show_followup = False",
    "            st.rerun()",
])

print("âœ… Rebuilt all files in pennyworth-chat/")
```
